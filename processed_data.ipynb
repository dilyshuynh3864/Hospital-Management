```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data/diabetic_data.csv")
df_clean = df.copy()
```

```python
df_clean = df_clean.replace(r'\?', np.nan, regex = True)
df_clean = df_clean.replace(r'^\s*$', np.nan, regex = True)
```

```python
df_clean.isna().sum()
```

```python
df_clean = df_clean.drop(['weight', 'payer_code'], axis = 1)
df_clean.shape
```

```python
df_clean['medical_specialty'] = df_clean['medical_specialty'].replace(np.nan, 'Unknown')
```

```python
df_clean['race'] = df_clean['race'].replace(np.nan, 'Other')
df_clean['race'].value_counts(dropna=False)
```

```python
df_clean = df_clean.dropna(subset = 'diag_1')
```

```python
df_clean['discharge_disposition_id'].value_counts()
```

```python
expired_hospice_id = [11, 13, 14, 19, 20, 21]
df_clean = df_clean[~df_clean['discharge_disposition_id'].isin(expired_hospice_id)]
df_clean.info()
```

```python
df_clean = df_clean.drop_duplicates(subset = ['patient_nbr'], keep = 'first')
df_clean.shape
```

```python
df_clean = df_clean[df_clean['diabetesMed'] == 'Yes']
df_clean.shape
```

```python
print(f'Number of rows before cleaning: {len(df)}')
print(f'Number of rows after cleaning: {len(df_clean)}')
```

```python
df_clean.to_csv('data/processed.csv', index = False)
```
